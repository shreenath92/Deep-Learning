```python
import re
import nltk
import pandas as pd
from tqdm import tqdm
from nltk.corpus import stopwords
from nltk.stem import  PorterStemmer
```

```python
import pandas as pd
df=pd.read_csv('/content/twitter_training.csv',header=None)
df=df.rename({0:'id',1:'comapny',2:'sentiment',3:'raw_tweet'},axis=1)
df=df.dropna()
df=df[df['sentiment']!='Irrelevant']
df=df[df['sentiment']!='Neutral']
```

```python
tweets=df['raw_tweet'].values
```

```python
proceesed_tweets=[]
for tweet in tqdm(tweets):
  tweet=re.sub(r'[^\w\s]','',tweet)
  tweet=tweet.lower()
  tweet=[word for word in tweet.split(' ')if not word in stopwords.words('english')]
  tweet=[stemmer.stem(word) for word in tweet]
  tweet=[word for word in tweet if len(word)!=0]
  tweet=' '.join(tweet)
  proceesed_tweets.append(tweet)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
enc=tfidf.fit_transform(proceesed_tweets)
df_=pd.DataFrame(enc.toarray(),columns=tfidf.get_feature_names_out())
df_
```

```python
x=enc.toarray()
y=pd.get_dummies(df["sentiment"]).values[:,1:]
```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(x,y)
```
